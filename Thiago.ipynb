# Thiago Alcântara do Nascimento

```python
import pandas as pd
import matplotlib.pyplot as plt
from qbstyles import mpl_style
import mplcyberpunk
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
allfifas=['players_15.csv','players_16.csv','players_17.csv','players_18.csv','players_19.csv','players_20.csv','players_21.csv']
c=1    
for i in allfifas:
    FIFA =pd.read_csv(i,index_col=False)
    FIFA.drop(["player_url","sofifa_id","short_name","age","dob","league_name","league_rank","preferred_foot","work_rate","body_type","real_face","team_jersey_number",],axis=1,inplace=True)
    Magic=FIFA[FIFA['long_name']=="Thiago Alcântara do Nascimento"]
    if c ==1:
        Magic.to_csv(r"Thiago\Thiago.csv",mode='w',index=False,header=True)
    else:
        Magic.to_csv(r"Thiago\Thiago.csv",mode='a',index=False,header=False)
    c+=1
```

Now I'm Cleaning the data coz there are some values that needs to be corrected


```python
Magico=pd.read_csv(r"Thiago\Thiago.csv",index_col=False)
columns=['ls','st','rs','lw','lf','cf','cam','lam','rw','rf','rb','rcb','cb','lcb','lb','rwb','rdm','cdm','ldm','lwb','rm','rcm','cm','lcm','lm','ram']
for pos in columns:
    Magico[pos]=Magico[pos].str.split("+")
    for (cell,i) in zip(Magico[pos],range(len(Magico[pos]))):
        cell=int(cell[0])+int(cell[1])
        Magico[pos][i]=cell
        if i==(len(Magico[pos])-1):
            break
Magico.to_csv("Thiago\Thiago repaired.csv",index=False,header=True)
Magico=pd.read_csv("Thiago\Thiago repaired.csv",index_col=False)
Magico.drop(["long_name","mentality_composure","nationality","club_name","player_positions","team_position","value_eur","loaned_from","joined","contract_valid_until","wage_eur","release_clause_eur","player_tags","nation_position","nation_jersey_number"],axis=1,inplace=True)
Magico.drop(["height_cm","weight_kg","international_reputation","gk_diving","gk_handling","gk_kicking","gk_reflexes","gk_speed","gk_positioning","player_traits","defending_marking"],axis=1,inplace=True)
```

```python
X=Magico.iloc[:,1:].values 
Y=Magico.iloc[:,0].values
# -----------------------
X_train,X_test,Y_train,Y_test=train_test_split(X , Y , test_size=0.25 , random_state=42 )
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
classifier=LinearRegression()
classifier.fit(X_train,Y_train) 
y_pred=classifier.predict(X_test)
print("Predicted rating for Thiago is",y_pred[0])
```

```python
years=[2015,2016,2017,2018,2019,2020,2021,2022]
Rating=Magico["overall"].to_list()
Rating.append(y_pred[0])
mpl_style("dark")
#plt.style.use("cyberpunk")

plt.plot(years, Rating,color="limegreen",marker='o')
plt.xlabel("Years")
plt.ylabel("Overall rating")
plt.title('Thiago Alcantara')
# mplcyberpunk.add_glow_effects(gradient_fill=True)
mplcyberpunk.make_lines_glow()
mplcyberpunk.add_underglow()
plt.savefig("Thiago\Thiago Rating with 2022 prediction.JPEG", dpi=300, bbox_inches='tight')
plt.show()
```
