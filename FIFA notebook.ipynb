# 2D bar chart for most common 5 Tags & every position
---

```python
Alltraits={}

Allpostions=[MidTags,DefTags,AttTags]
for postion in Allpostions:
    for Tag in postion:
        if Tag in Alltraits:
            occ=Alltraits[Tag]
            occ=occ+postion[Tag]
            Alltraits.update({Tag:occ})
        else:
            Alltraits.update({Tag:postion[Tag]})

        
Alltraits={k: v for k, v in sorted(Alltraits.items(), key=lambda item: item[1])}
Alltraits_list=list(Alltraits.items())
Alltraits_list.pop()
required_traits=[]
for x in range(-5,0):
    required_traits.append(Alltraits_list[x][0])
midfreq=[]
deffreq=[]
attfreq=[]
postions_freq=[midfreq,deffreq,attfreq]
print(required_traits) ## this is the list of the most common tag among the postions
for (postion,postion_freq) in zip(Allpostions,postions_freq):
    for trait in required_traits:
        for Tag in postion:
            if  Tag==trait:
                postion_freq.append(postion[Tag])
mpl_style("dark")
print(attfreq)
plt.bar(np.arange(len(attfreq))+0.2, attfreq,width=0.2,align="center",label="Attackers",color="red")
plt.bar(np.arange(len(midfreq)), midfreq,width=0.2,align="center",label="Midfielders",color="limegreen")
plt.bar(np.arange(len(deffreq))+0.4, deffreq,width=0.2,align="center",label="Defenders",color="blue")
plt.xticks(np.arange(len(midfreq)),required_traits)

plt.xlabel("Most 5 common skills")
plt.ylabel("Frequency")
plt.title('Skills among postions')
mplcyberpunk.add_glow_effects(gradient_fill=True)
plt.legend(loc='best')

plt.savefig("Analysis\Skills among positions.JPEG", dpi=300, bbox_inches='tight')
plt.show()
```
